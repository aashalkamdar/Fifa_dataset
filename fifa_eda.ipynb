```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('FIFA 2018 Statistics.csv')
```

```python
df
```

```python
df['Shot Accuracy in %'] = ((df['On-Target'])/(df['On-Target']+df['Off-Target']))*100
```

-  calculated the accuracy of a teams shots so that on target and off target can be removed


```python
df['Conversion Rate'] = (df['Goal Scored']/df['Attempts'])*100

```

-  conversion rate is the total number of goals scored by a team for their total attempts 

```python
df.drop('Attempts',inplace=True,axis=1)
```

-  we can drop attempts because it became a subset of conversion rate

```python
df.drop('On-Target',inplace=True,axis=1)
```

```python
df.drop('Off-Target',inplace=True,axis=1)
```

-  on target and off target both can be dropped as they became a subset of shot accuracy

```python
pivot = df.pivot_table(index='Team',columns='Opponent')
```

```python
df.drop('Passes',axis=1,inplace=True)
```

```python
df.drop('Yellow & Red',axis=1,inplace=True)
```

-  we already have number of yellow and red cards individually so the sum isnt required

```python
df.drop('Man of the Match',axis=1,inplace=True)
```

-  not important in predicting the outcome because we already have the goals scored and so the winner can be found

```python
df.drop('Own goals',axis=1,inplace=True)
```

-  we already have total goal so not required

```python
df.drop('Own goal Time',axis=1,inplace=True)
```

-  if own goal isn't required we do not need own goal time

```python
df.drop('Saves',axis=1,inplace=True)
```

```python
df.drop('Blocked',axis=1,inplace=True)
```

-  does not help in predicting the winner because we have goal scored

```python
df.fillna(value='did not score',inplace=True)
```

```python
pivot.fillna('Did not play')
```

```python
t=df.groupby('Team')['Goal Scored'].sum()
```

```python
df.drop('Distance Covered (Kms)',axis=1,inplace=True)
```
